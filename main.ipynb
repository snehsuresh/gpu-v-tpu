```python
import json
import matplotlib.pyplot as plt

with open('training_history_gpu.json', 'r') as file:
    data1 = json.load(file)

# Load data from file 2
with open('training_history_tpu.json', 'r') as file:
    data2 = json.load(file)


plt.plot(data1['accuracy'], color='red', label='GPU - Training', linestyle='dotted')
plt.plot(data2['accuracy'], color='blue', label='TPU - Training')


plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Show the plot
plt.show()
```

```python
plt.plot(data1['val_accuracy'], color='red', label='GPU - Validation', linestyle='dotted')
plt.plot(data2['val_accuracy'], color='green', label='TPU - Validation')

# # Training times
# plt.text(0.5, 0.9, f'Training Time (GPU): ~400 seconds', transform=plt.gca().transAxes, color='red')
# plt.text(0.5, 0.85, f'Training Time (TPU): ~90 seconds', transform=plt.gca().transAxes, color='blue')

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Show the plot
plt.show()
```
